# Problema #1: Calculadora de propina inteligente

## Problema

- Calcular el total de la cuenta e incluir propina

## Entrada
- Total de la cuenta
- Porcentaje propina 

## Proceso
- Caluclar la propina y sumarla a la cuenta final


## Salida esperada

- Tener el total a pagar y la propina, y si ingresa datos invalidos, le de la opcion de colocar datos correctos utilizando el tryexcept

## Pruebas manuales

| Escenario | Datos de entrada | Resultado esperado | Resultado obtenido |
|-----------|------------------|--------------------|--------------------|
| Caso normal | total 10000 propina 10%| 11000| 11000|
| Caso límite | total 0 propina 10| 0| 0|
| Caso inválido |total"hola" propina 10 | mensaje error tryexcept | colocar nuevamente datos numericos |


```python
# Escriba la solución aquí
try:

    total = float(input("total de la cuenta: ")) 
    porcentaje = float(input("porcentaje propina: "))
    
    propina = total * porcentaje / 100
    total_final = total + propina
    print(f"propina: {propina}")
    print(f"total a pagar: {total_final}")
    
except ValueError:
    print("Error: ingresa numeros validos")
```

# Problema #2: Detector de número par/impar

## Problema
Detector de número par/impar 

## Entrada

- Solicitar numero entero

## Proceso

- convertir numero entero
- calucula el porcentaje

## Salida esperada

- - identificar si es numero par o impar


## Pruebas manuales

| Escenario | Datos de entrada | Resultado esperado | Resultado obtenido |
|-----------|------------------|--------------------|--------------------|
| Caso normal |10 | par|par|
| Caso límite | 0|impar |impar |
| Caso inválido |diez | error|error: Ingresa numero |


```python
# Escriba la solución aquí
try:
    numero = int(input("Ingresa numero: "))
    if numero % 2 == 0:
        print("par")
    else:
        print("impar")
except ValueError:
    print("Error: Ingresa numero")
```

# Problema #3: Clasificador de edad

## Problema

 - Solicitar edad y clasifica como menor, adulto o adulto mayor. 

## Entrada

- Edad de cada persona

## Proceso

- menor a 18 años (menor de edad) 
- 18 a 64 (adulto) 
- 65 a mas (adulto mayor)

## Salida esperada

- Mostrar que clasificacion tiene la persona (menor de edad - adulto - adulto mayor)

## Pruebas manuales

| Escenario | Datos de entrada | Resultado esperado | Resultado obtenido |
|-----------|------------------|--------------------|--------------------|
 Caso normal | 18 | adulto | adulto |
| Caso límite | 65| adulto mayor|adulto mayor |
| Caso inválido |veinte |error | Solicita datos numericos|


```python
# Escriba la solución aquí

try:
    edad = int(input("ingrese su edad: "))
    
    if edad < 18:
        print("Menor de edad")
    elif edad < 65:
        print("Adulto")
    else:
        print("Adulto Mayor")
        
except ValueError:
    print("Error: Ingrese datos numericos")    
    
    
```

# Problema #4: Validador de compra con descuento

## Problema

Validador de compra con descuento, segun el monto y si el cliente es frecuente
## Entrada

- Nombre del cliente
- Monto de la compra
- Cliente frecuente (Si/No)

## Proceso
- ingresar datos del cliente
- si es cliente frecuente aplica descuento

## Salida esperada

- nombre
- monto y descuento
- total a pagar

## Pruebas manuales

| Escenario | Datos de entrada | Resultado esperado | Resultado obtenido |
|-----------|------------------|--------------------|--------------------|
| Caso normal | Pamela, 10000, si | descuento 500 | total 9500 |
| Caso límite | Pamela, 10000, no| sin descuento|total a pagar |
| Caso inválido |Pamela, texto, si |error | error: colocar numero|


```python
# Escriba la solución aquí
try:
    cliente = input("Nombre cliente: ")
    monto = float(input("Ingresar el monto: "))
    frecuente = input("Es cliente frecuente: ").strip().lower()
    
    if monto >= 10000 and frecuente == "si":
     descuento = monto * 0.05
    
    else:
        descuento = 0
    
    total = monto - descuento
    print(f"Cliente: {cliente}")
    print(f"Descuento: {descuento}")
    print(f"total: {total}")

except ValueError:
    print("Error: Monto en numero")
```
